```
import os
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
import pickle as pkl
from sklearn.decomposition import PCA
```

# Biker going/not going

```
tour_convoy = pd.read_csv('./dataset/tour_convoy.csv')
tour_convoy.index = tour_convoy['tour_id']
tour_convoy.drop(['tour_id'], axis=1, inplace=True)
tours_dict = tour_convoy.index
display(tour_convoy)
```

```
bikers_dict = pkl.load(open('./Required Data/bikers_list.pkl', 'rb'))
tours_dict = pkl.load(open('./Required Data/tours_list.pkl', 'rb'))
```

```
biker_tours = np.zeros((len(bikers_dict.keys()), len(tours_dict.keys())))
for tour_id in tqdm(tours_dict.keys()):
    going = tour_convoy.loc[tour_id]['going']
    not_going = tour_convoy.loc[tour_id]['not_going']
    maybe = tour_convoy.loc[tour_id]['maybe']

    try:
        going = going.split()
        for biker_id in going:
            try:
                biker_tours[bikers_dict[biker_id], tours_dict[tour_id]] = 1
            except:
                continue
    except:
        continue

    try:
        not_going = not_going.split()
        for biker_id in not_going:
            try:
                biker_tours[bikers_dict[biker_id], tours_dict[tour_id]] = -1
            except:
                continue
    except:
        continue

    try:
        maybe = maybe.split()
        for biker_id in maybe:
            try:
                biker_tours[bikers_dict[biker_id], tours_dict[tour_id]] = 0.5
            except:
                continue
    except:
        continue
```

```
tour_biker_data = pd.DataFrame(data=biker_tours, index=bikers_dict.keys(), columns=tours_dict.keys())
display(tour_biker_data)
```

```
tour_biker_data.to_csv('./Required data/tour-biker.csv')
```
